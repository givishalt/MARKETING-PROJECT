### 3. Experiment design

This part of the project is focused on the experimental design process and hypothesis testing. Based on evaluation question we will formulate and test a hypothesis using Spreadsheets.

Marketing studies begin with an evaluation question, which may come from many sources - manager, stakeholder, client or even yourself.

For this project the **evaluation question** is as follows 

**"Is there a difference between the number of conversions on Facebook platform vs Adwords platform?"**

The full experimental design process will involve five basic steps:

1. Questioning
2. Hypothesis
3. Required Variables
4. Choosing a Measurement Approach
5. Selecting an Analysis


### 3.1 Questioning and Hypothesis

The first task we want to address is to write a hypothesis statement that answers the evaluation question. To do that we will need to andwer the following three questions that will help us to formulate a good hypothesis:

1. What will change?
2. How will it change?
3. What will cause the change?

Based on that our hypothesis will be:

**"Our number of conversions will be greater if we advertise on platform Facebook rather than platform AdWords"**

### 3.2 Required Variables

In a test like this, we call the variable that we manipulate an independent variable. The variable whose effect we measure is called a dependent variable. Therefore, we consider variables as follows:

**Independant variable:** platform

**Dependant variable:** number of conversions


### 3.3 Testing Hypotesis

By answering the type of questions like "is there a difference" we can receive two possible answers or hypotheses:

**(H0)** There is no difference in conversions

**(H1)** The number of conversions on Facebook are higher than the number of conversions on AdWords

Hypothesis tests are designed to determine which of these hypotheses—the null hypothesis or the alternative hypothesis—is the true one.

Here are the things we will do to test the hypothesis:

a. Calculate the mean values for the “Facebook Ad Conversions” and “AdWords Ad Conversions” with the =AVERAGE formula

b. Next, we will determine whether this overall difference is a statistically significant difference by comparing the p-value for the data to some given Alpha value (𝛂), which is 5% in this case (or 0.05). 

We can perform the test using formula = TTEST(A:A,B:B,2,2)



Results under the value of alpha (0.05) imply high significance, therefore, p-value of 0.00 implies high significance. In this case  we reject H0 and accept H1. Which means that we will receive more conversions if we advertise on Facebook platform rather than on AdWords
